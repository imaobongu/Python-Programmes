#### 1. Using read_csv(), load the SP500_Constituents_20230101.csv file into a DataFrame. Use the data in the first column (Symbol) as the index of the data.

```python
# read data from the csv file into a dataframe
filename = "SP500_Constituents_20230101.csv"
df = pd.read_csv(filename,
                index_col = "Symbol")
```

#### 4. Which columns have missing values?

```python
# determine which columns have missing values.
# output shows "missing values in column" is true for 'Date first added' column
df.isnull().any()
```

```python
# get data for the 'AAPL' row as a series
df.loc["AAPL"]
```

#### 6. What are the distinct values within the column 'GICS Sector'?

```python
# display unique values for the 'GICS Sector' column
df["GICS Sector"].unique()
```

#### 7. How many stocks are there in each of the GICS Sectors?

```python
# show count of stocks for each unique 'GICS Sector' value
df["GICS Sector"].value_counts()
```

#### 8. Use a filter expression to display the rows where the stock is in the Energy sector.

```python
# create and display df with only rows that have 'GICS Sector' as Energy
energydf = df[df["GICS Sector"] == "Energy"]
energydf
```

#### 9. Use sort_values() to order to data for Energy stocks, sorting based on the 'Date first added'. What company was added first, what company was added most recently?

```python
# sort energy df from earliest to most recent
# output shows first company added is Halliburton
# output shows most recent company added is Targa Resources
# valero energy has no date identified
energydf.sort_values("Date first added", ascending=True)
```
